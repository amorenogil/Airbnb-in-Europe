## Creating a correlation heatmap

```python
#creating a correlation matrix
sub.corr()
```

```python
#creating a subplot with matplotlib
f,ax = plt.subplots(figsize=(9,9))

#creating the correlation map in seaborn
corr = sns.heatmap(sub.corr(), annot = True, ax=ax)
```

**Most of the strong correlations that appear in the heatmap are quite self-evident, for example, it is very logical that the variables number_of_review, reviews_per_month and number_of_reviews_ltm (last 12 months) are strongly correlated. Consequently the fact that they are correlated doesn't explain much, as it would be rare if they weren't.**

**Perhaps the most important variable to analyze for my study is the "price" as I'm particularly interested in how and why does the prize of Airbnb listings fluctuate. In this case we see that the price is mostly not correlated to other variables, such as the review variables or the minimum nights. THe strongest variables correlated with the price is latitude (0.14); this indicates that price varies primarily based on an N-S geographical axis. The are also seems to be a relation between the geographical location and the availability of the accomodations (0.31 for latitude and 0.19 for longitude). Finally, there seems to also be a weak correlation between the review-variables and the availability.**

**Some interesting information can be obtained from the categorical graphic. Other visible trends are cities that have less availability of cheap listings, e.g. Amsterdam, Geneva, Rotterdam, The Hague and Zurich.** 

# 04. Answering questions and hypothesis

## Questions

**Are there significant differences in the pricing of the Airbnb accomodations across the different cities of Europe?**

Yes, some differences can be observed,  especially with regard to high price listings, which concentrate in certain cities, such as London, Paris, Florence or Rome and are much scarcer in other like Riga, or Ghent. Similarly cheap accomodations are more abundant in certain cities such as Istanbul, Riga, Porto, Barceloa or Berlin. 

**The price paid for an Airbnb will outweigh, by far, the average rent paid by locals**

This is definetely true, using Barcelona as an example the average rent in the city in december 2022 was 1066 euros (source: https://www.eldiario.es/catalunya/precios-alquiler-barcelona-baten-propio-record-establecen-1-066-euros-media_1_9812444.html), whereas the average price paid per night for an airbnb was 103 euros (see infra), which means that  per month the cost is upwards of 3090 euros, i.e. it triples the normal rent of the city

```python
city_groups = df_bnb.groupby('city')
```

```python
city_means = city_groups.mean()
```

```python
print ('The average Airbnb price for Barcelona is', city_means.loc['Barcelona','price'],'euros')
```
